```python
# 8. Create a 2d array with 1 on border and 0 inside
def bordered(n, m):
    A = np.ones((n,m), dtype=int)
    if n>2 and m>2:
        A[1:-1,1:-1] = 0
    elif n>2 and m<=2:
        A[1:-1,:] = 0
    elif m>2 and n<=2:
        A[:,1:-1] = 0
    return A

bordered_6x6 = bordered(6,6)
bordered_6x6
```

```python
# 1. Prove A * A^{-1} = I (matrix inverse)
A = np.array([[3,4],[7,8]], dtype=float)
A_inv = np.linalg.inv(A)
I_calc = A.dot(A_inv)
I_calc_rounded = np.round(I_calc, decimals=10)
A_inv, I_calc_rounded
```

```python
# 2. Show AB != BA (matrix multiplication is not generally commutative)
AB = A.dot(B)
BA = B.dot(A)
AB, BA, np.array_equal(AB, BA)
```

```python
# 3. Prove (AB).T = B.T @ A.T
lhs = (A.dot(B)).T
rhs = B.T.dot(A.T)
lhs, rhs, np.array_equal(lhs, rhs)
```

```python
# Solve system of linear equations using matrix form (AX = B)
# Equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9
A_mat = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]], dtype=float)
B_vec = np.array([-1, -3, 9], dtype=float)
solution = np.linalg.solve(A_mat, B_vec)
solution
```

```python
# Solve same system using inverse (np.linalg.inv) to demonstrate
A_inv = np.linalg.inv(A_mat)
solution_using_inv = A_inv.dot(B_vec)
solution_using_inv
```

```python
import time
import random

N = 1_000_000

# Prepare data
list_a = [random.random() for _ in range(N)]
list_b = [random.random() for _ in range(N)]

# 1. Element-wise addition - lists
t0 = time.time()
list_add = [a + b for a, b in zip(list_a, list_b)]
t1 = time.time()
time_list_add = t1 - t0

# 1. Element-wise addition - numpy
arr_a = np.array(list_a)
arr_b = np.array(list_b)
t0 = time.time()
np_add = arr_a + arr_b
t1 = time.time()
time_np_add = t1 - t0

# 2. Element-wise multiplication - lists
t0 = time.time()
list_mul = [a * b for a, b in zip(list_a, list_b)]
t1 = time.time()
time_list_mul = t1 - t0

# 2. Element-wise multiplication - numpy
t0 = time.time()
np_mul = arr_a * arr_b
t1 = time.time()
time_np_mul = t1 - t0

# 3. Dot product - lists
t0 = time.time()
dot_list = sum(a*b for a,b in zip(list_a, list_b))
t1 = time.time()
time_list_dot = t1 - t0

# 3. Dot product - numpy (using np.dot)
t0 = time.time()
dot_np = arr_a.dot(arr_b)
t1 = time.time()
time_np_dot = t1 - t0

# 4. Matrix multiplication: pure-python smaller size, numpy 1000x1000
size_small = 200  # smaller size for pure Python to keep runtime reasonable here
# pure-Python matrix multiplication (size_small x size_small)
A_list = [[random.random() for _ in range(size_small)] for _ in range(size_small)]
B_list = [[random.random() for _ in range(size_small)] for _ in range(size_small)]

def matmul_list(A, B):
    n = len(A)
    m = len(B[0])
    p = len(B)
    C = [[0.0]*m for _ in range(n)]
    for i in range(n):
        Ai = A[i]
        for k in range(p):
            Bk = B[k]
            aik = Ai[k]
            rowC = C[i]
            for j in range(m):
                rowC[j] += aik * Bk[j]
    return C

t0 = time.time()
C_list = matmul_list(A_list, B_list)
t1 = time.time()
time_list_matmul = t1 - t0

# numpy matrix multiplication 1000x1000
size_big = 1000
A_np_big = np.random.random((size_big, size_big))
B_np_big = np.random.random((size_big, size_big))
t0 = time.time()
C_np_big = A_np_big.dot(B_np_big)
t1 = time.time()
time_np_matmul = t1 - t0

# Summarize times (seconds)
times = {
    'list_add': time_list_add,
    'numpy_add': time_np_add,
    'list_mul': time_list_mul,
    'numpy_mul': time_np_mul,
    'list_dot': time_list_dot,
    'numpy_dot': time_np_dot,
    'list_matmul_{}x{}'.format(size_small, size_small): time_list_matmul,
    'numpy_matmul_{}x{}'.format(size_big, size_big): time_np_matmul
}
times
```

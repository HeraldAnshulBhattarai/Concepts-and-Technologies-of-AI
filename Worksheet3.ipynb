```python
# LOAD DATASET
import pandas as pd
import numpy as np
path='/content/drive/MyDrive/Concept and technologies of AI/week3/Copy of Titanic-Dataset.csv'
df=pd.read_csv(path)
print(df.info()); df.head()
```

```python
# COLUMN DETECTION
def find_col(df,candidates):
    for cand in candidates:
        for col in df.columns:
            if cand.lower() in col.lower(): return col
    raise KeyError()
fare_col=find_col(df,['fare']); pclass_col=find_col(df,['pclass','class']); age_col=find_col(df,['age']); survived_col=find_col(df,['survived']); sex_col=find_col(df,['sex']); embarked_col=find_col(df,['embarked']); name_col=find_col(df,['name']);
fare_col, pclass_col, age_col, survived_col, sex_col, embarked_col, name_col
```

```python
# Q2.3 female under 18
df[sex_col]=df[sex_col].astype(str).str.lower(); df[(df[age_col]<18)&(df[sex_col]=='female')].head()
```

```python
# Q2.4 embarked C or S
df[embarked_col]=df[embarked_col].astype(str).str.upper(); df[df[embarked_col].isin(['C','S'])].head()
```

```python
# Q2.5 pclass in [1,2]
df[df[pclass_col].isin([1,2])].head()
```

```python
# Fill age median
age_median=df[age_col].median(); df[age_col]=df[age_col].fillna(age_median); age_median
```

```python
# fare_per_year
df['fare_per_year']=df[fare_col]/df[age_col]; df['fare_per_year'].head()
```

```python
# subset fare_per_year>5
df[df['fare_per_year']>5].head()
```

```python
# sort fare_per_year
df[df['fare_per_year']>5].sort_values(by='fare_per_year',ascending=False).head()
```

```python
# select name & fare_per_year
df[df['fare_per_year']>5].sort_values(by='fare_per_year',ascending=False)[[name_col,'fare_per_year']].head(10)
```

```python
# highest fare_per_year
df[df['fare_per_year']>5].sort_values(by='fare_per_year',ascending=False)[[name_col,'fare_per_year']].iloc[0]
```

```python
# fare_per_class
df['fare_per_class']=df[fare_col]/df[pclass_col]; df['fare_per_class'].head()
```

```python
# adult males
df[(df[sex_col]=='male')&(df[age_col]>=18)].head()
```

```python
# sort adult males
df[(df[sex_col]=='male')&(df[age_col]>=18)].sort_values(by='fare_per_class',ascending=False).head()
```

```python
# select cols
df[(df[sex_col]=='male')&(df[age_col]>=18)].sort_values(by='fare_per_class',ascending=False)[[name_col,age_col,'fare_per_class']].head(10)
```

```python
# top adult male
df[(df[sex_col]=='male')&(df[age_col]>=18)].sort_values(by='fare_per_class',ascending=False)[[name_col,age_col,'fare_per_class']].iloc[0]
```

```python
# total fare
df[fare_col].sum()
```

```python
# fare per class
df.groupby(pclass_col)[fare_col].sum()
```

```python
# age groups
def grp(a): return 'child' if a<18 else ('adult' if a<=64 else 'senior')
df['age_group']=df[age_col].apply(grp); df['age_group'].value_counts()
```

```python
# count by age group
df['age_group'].value_counts()
```

```python
# percentage age group
df['age_group'].value_counts()/len(df)*100
```
